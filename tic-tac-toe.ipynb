```python
import random

board2 = [['*' for i in range(3)] for j in range(3)]

winningCombos = (
    [(0, 0), (0, 1), (0, 2)], [(1, 0), (1, 1), (1, 2)], [(2, 0), (2, 1), (2, 2)],
    [(0, 0), (1, 0), (2, 0)], [(0, 1), (1, 1), (2, 1)], [(0, 2), (1, 2), (2, 2)],
    [(0, 0), (1, 1), (2, 2)], [(0, 2), (1, 1), (2, 0)]
)


def printBoard(board2):
    for i in range(len(board2)):
        print(board2[i][0], board2[i][1], board2[i][2])


initialPos = random.randint(0, 8)
col = initialPos % 3
if (initialPos > 2 and initialPos < 6):
    board2[1][col] = 'X'
elif initialPos > 2 and initialPos > 5:
    board2[2][col] = 'X'
else:
    board2[0][col] = 'X'


def availablePos():
    return ([(i, j) for i in range(3) for j in range(3) if board2[i][j] == '*'])


def getOpponent():
    if len(availablePos()) % 2 == 0:
        return 'O'
    return 'X'


def playMove(pos, turn):
    board2[pos[0]][pos[1]] = turn


def getWinner():
    for player in ["X", "0"]:
        for rule in winningCombos:
            if board2[rule[0][0]][rule[0][1]] == player and board2[rule[1][0]][rule[1][1]] == player and \
                    board2[rule[2][0]][rule[2][1]] == player:
                return player
    if len(availablePos()) == 0:
        return 'tie'
    return None


def start():
    if len(availablePos()) == 0:
        return True
    if getWinner() != None:
        return True
    return False

def printwinner():
    if getWinner() == 'X':
        return "X wins"
    elif getWinner() == 'O':
        return "O wins"
    elif getWinner() == 'tie':
        return "Draw Game"



def algo(turn, depth=0):
    if turn == 'O':
        bestVal = -10
    else:
        bestVal = 10

    if start():
        if getWinner() == 'X':
            return -10 + depth, None
        elif getWinner() == 'O':
            return 10 - depth, None
        elif getWinner() == 'tie':
            return 0, None
        # return getWinner()
    for pos in availablePos():
        # printBoard(board2)
        playMove(pos, turn)
        # print('------------------------')
        value, _ = algo(getOpponent(), depth + 1)
        playMove(pos, '*')

        if turn == 'O':
            if value > bestVal:
                bestVal, bestPos = value, pos
        else:
            if value < bestVal:
                bestVal, bestPos = value, pos

    return bestVal, bestPos
    # printBoard(board2)


printBoard(board2)
# availablePos()

for i in range(len(availablePos())):
    player = getOpponent()
    #print(player)
    bestLoc = algo(getOpponent())[1]
    #print(bestLoc)
    playMove(bestLoc, player)
    print('-----')
    printBoard(board2)
print("Result: ", printwinner())



```
